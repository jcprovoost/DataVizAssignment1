# Data Visualization Assignment 1

From the Worldbank Data, I have chosen the dataset about access to electricity. Firstly, I had to wrangle and clean this dataset since the data was not properly formatted. Also, there were a lot of missing (NaN) values. I did this using the Trifacta Wrangler software. The wrangled dataset was exported in CSV format and loaded into a Pandas DataFrame.

```python
import pandas as pd

data = pd.read_csv('data.csv')
```

Afterwards, I used the Bokeh framework to give an insight into the development of electricity access per country. When the country is selected from the menu, the corresponding graph is shown. 
## Electricity access per country

```python
from bokeh.io import push_notebook, show, output_notebook
from bokeh.plotting import figure
from bokeh.models import Select, ColumnDataSource
from ipywidgets import interact, Dropdown

output_notebook()

source = ColumnDataSource(data)
country_list = data["Country_Name"].dropna().unique().tolist()

p = figure(x_axis_label='Year', y_axis_label='Access to electricity (%)')
l = p.line("year", "electricityAccess", source=source, color="blue", line_width=3, line_alpha=0.5)

country_select = Select(title="Country:", options=country_list)

def update(f):
    filtered_data = data[data["Country_Name"]==f]
    l.data_source.data = ColumnDataSource.from_df(data[data["Country_Name"] == f])
    push_notebook()

show(p, notebook_handle = True)

menu = Dropdown(
    options=country_list,
    value=country_list[0],
    description="Country:"
)

interact(update, f=menu)
```
